# 41. Line plot with multiple lines and shaded the area :-

```python
average_duration_by_date_gender = df.groupby(['Date', 'Gender'])['Visit Duration'].mean().unstack()
std_duration_by_date_gender = df.groupby(['Date', 'Gender'])['Visit Duration'].std().unstack()
average_duration_by_date_gender.plot()
plt.fill_between(average_duration_by_date_gender.index, average_duration_by_date_gender['Female'] - std_duration_by_date_gender['Female'],
                 average_duration_by_date_gender['Female'] + std_duration_by_date_gender['Female'], alpha=0.3)
plt.fill_between(average_duration_by_date_gender.index, average_duration_by_date_gender['Male'] - std_duration_by_date_gender['Male'],
                 average_duration_by_date_gender['Male'] + std_duration_by_date_gender['Male'], alpha=0.3)
plt.xlabel('Date')
plt.ylabel('Average Visit Duration')
plt.title('Average Visit Duration over Time by Gender')
plt.legend(title='Gender')
plt.show()
```
